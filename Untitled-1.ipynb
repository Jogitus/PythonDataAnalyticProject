<h4>Handling Missing Values</h4>
<p>You can see that the missing values in the dataset I have chosen are specific moves in the mainlines.</p>
<p>Main line means the theory of an opening that should be played, but due to nature of chess and freedom of moves.</p>
<p>If I play pawn to D4, my opponent can usually expect London System or the Queens Gambit.</p>
<p>But that also means, they cant use a opening they would play against pawn to e4 (Such as Caro-Kann or the Sicilian defence)</p>
<p>So basically the missing values are just moves we dont make in the mainlines.</p>
<p>For example move1b (So blacks first move in the game.) We have 15 empty cells.</p>
<p>That means that in 15 of our openings, we do not have a mainline response to whites opening move.</p>
<p>In the codebox below, I have used df.fillna(" "), to replace all null values with " " string. <br> Meaning they are still empty, but we can work with them.</p>
<p> </p>
<p> Please note that the codebox with the missing values is the dataset with nothing done to it.<br> It is there, so we can see that we infact have null data</p>

<h4>Calculating Mean & Median</h4>
<p>Here I chose to calculate the Mean & the Median for the average ELO of a player.</p>
<p>In my opinion this was the best Column to do so.</p>
<p>The data result we get is really interessting, and we get an idea of how good the players actually are in this dataset.</p>

```python
mean = df['Avg Player'].mean()
median = df['Avg Player'].median()

print("Mean of Average Players ELO: {:.1f}".format(mean))
print("Median for Average Players ELO: {:.0f}".format(median))
```

<h4>Calculating Standard Deviation & Variance</h4>
<p>Again we are using Avg Player for this example.</p>
<p>From here we can see the standard deviation, so we know it spreads out 127.7 from our mean.</p>
<p>We can also see that our Variance is quite high at 16313.3</p>

```python
deviation = df['Avg Player'].std()
variance = df['Avg Player'].var()

print("Deviation of Average Player: {:.1f}".format(deviation))
print("Variance of Average Player: {:.1f}".format(variance))
```

<h4>Calculating the minimum & the maximum</h4>
<p>So once again I have chosen to use Average ELO of players here.</p>
<p>We see that the Minimum ELO is 1577 & Maximum ELO is 2492</p>
<p>Considering that the Grandmaster(GM) title is "unlocked" when you reach 2500 ELO in official tournaments.</p>
<p>We can see that from our Dataset my quess is we have few International-masters (IM) here, and out top player closing on that GM-Title.</p>
<p>(Fun fact, highest recorded ELO rating is 2882, achieved by Magnus Carlsen in 2014.)</p>

```python
Minimum = np.min(df['Avg Player'])
Maximum = np.max(df['Avg Player'])
print("The Minimum ELO of a player is: ", Minimum)
print("The Maximum ELO of a player is: ", Maximum)
```

<h4>Bar Chart & Histogram</h4>
<p>The bar chart shows the amount of games for average ELO of players</p>
<p>We can see that the amount of games is higher with higher ELO.</p>
<p>This effect comes from playing harder opponents, so most likely you need to play more games to gain ELO.</p>
<p>The Histogram shows how many players belong to each ELO "bracket"</p>
<p>Here we can see that most of the players in this dataset belong to the 2200 ELO range.</p>

```python
# BARCHAR
x1 = df['Avg Player']
y1 = df['Num Games']
plt.bar(x1, y1, label="Blue Bar", color='b')
plt.xlabel("Player ELO")
plt.ylabel("Games Played")
plt.title("Amount of games for ELO")
plt.legend()
plt.show()

# HISTOGRAM
data = df['Avg Player']
plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('ELO of the player')
plt.ylabel('Amount of players in said ELO')
plt.title('ELO of players')
plt.show()
```

<h4>Pie Chart & Heatmap</h4>
<p>The pie chart tells us the outcome % of the played games.</p>
<p>We see that white pieces have a clear advantage here. (They have the first move, so they can dictate the tempo of the game)</p>
<p>What is interessting to see is that draw is almost the same as black winning.</p>
<p>(Compered to other "sports" draw in professional play is really really common amongst the best players.)</p>
<p>It is usually said that when chess is played perfectly from both players the outcome will always be a draw.</p>

```python
# Pie
labels = 'Black_Win%', 'Draw%', 'White_Win%'
a = df['Black_win%'].median()
b = df['Draw %'].median()
c = df['White_win%'].median()
sizes = [a, b, c]
colors = ['c', 'g', 'y']
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%')
plt.axis('equal')
plt.title('% of the games outcome')
plt.show()

# Heatmap
```
