```python
# Mean and Median (0.5 points): Correctly calculates the mean and median values for appropriate columns.
mean = df['Avg Player'].mean()
median = df['Avg Player'].median()

print("Mean of Average Players ELO: {:.1f}".format(mean))
print("Median for Average Players ELO: {:.0f}".format(median))
```

```python
# Standard Deviation and Variance (0.5 points): Accurately computes the standard deviation and variance of data.
deviation = df['Avg Player'].std()
variance = df['Avg Player'].var()

print("Deviation of Average Player: {:.1f}".format(deviation))
print("Variance of Average Player: {:.1f}".format(variance))
```

```python
# Minimum and Maximum (0.5 points): Finds the minimum and maximum values.
Minimum = np.min(df['Avg Player'])
Maximum = np.max(df['Avg Player'])
print("The Minimum ELO of a player is: ", Minimum)
print("The Maximum ELO of a player is: ", Maximum)
```

```python
# Sum and Product (0.5 points): Accurately calculates the sum and product of data.
total_games = np.sum(df['Num Games'])
product_games = np.prod(df['Num Games' ])
print("The Sum of total amount of games is: ", total_games)
print("The Product of total amount of games is: ", product_games)
```

```python
# Line Plot and Scatter Plot (0.5 points): Creates both a line plot and a scatter plot with appropriate labels and titles.

#LINE PLOT
plt.plot(df['Avg Player'])
plt.xlabel('Players')
plt.ylabel('ELO')
plt.title('Visual of Players Average ELO')
plt.show()


# SCATTER PLOT
y1 = df['Avg Player']
x1 = df['Num Games']
plt.scatter(x1, y1)
plt.title('Scatter Plot Example')
plt.show()
```

```python
# Bar Chart and Histogram (0.5 points): Successfully generates a bar chart and a histogram.

# BARCHAR
x1 = df['Avg Player']
y1 = df['Num Games']
plt.bar(x1, y1, label="Blue Bar", color='b')
plt.xlabel("Player ELO")
plt.ylabel("Games Played")
plt.title("Amount of games for ELO")
plt.legend()
plt.show()

# HISTOGRAM
data = df['Avg Player']
plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('ELO of the player')
plt.ylabel('Amount of players')
plt.title('ELO of players')
plt.show()
```
