```python
# Saamme listattua uuden kategorian taulukkoomme jota kautta voimme saada haluttua statistiikkaa
# Esimerkiksi onko tuote kallis/halpa, taito tasoa tai muuta vastaavaa.

def categorize_MW(Black_Wins):
    if Black_Wins < 1000:
        return 'Noob'
    elif Black_Wins < 1700:
        return 'decent'
    else:
        return 'ProMan'

df['mustan taito'] = df['Black_Wins'].apply(categorize_MW)
df.head(5)
```
